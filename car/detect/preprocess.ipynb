```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('./bb.jpeg')
plt.imshow(img)
```

```python
import numpy as np
lane_img = np.copy(img)
gray = cv2.cvtColor(lane_img, cv2.COLOR_RGB2GRAY)
# cv2.imshow('gray',gray)
# cv2.waitKey(0)
plt.imshow(gray)
```

```python
blur=cv2.GaussianBlur(gray,(9,9),0)
plt.imshow(blur)
```
